Exploratory Data Analysis

```python
ax=sns.countplot(x="Gender" , data=df, saturation=0.2 , palette="Accent")
plt.title("Sum Of Sales By Gender")

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen=df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount" , ascending=False)
sns.set(style="darkgrid")
sns.barplot(x="Gender" , y="Amount" , data=sales_gen , palette=["Red", "Green"] , hue="Gender" ,
            width=0.6  , saturation=0.4 ,  )
plt.title("Number of sales on the basis of Gender")

# Customize the legend
legend = plt.legend(title="Gender", labels=["Female", "Male"])
# Set the colors, font weight, and font size for the legend text
legend.get_texts()[0].set_color("Red")    # Set color for "Female"
legend.get_texts()[0].set_fontweight("bold")  # Set font weight for "Female"
legend.get_texts()[0].set_fontsize(12)  # Set font size for "Female"

legend.get_texts()[1].set_color("Green")  # Set color for "Male"
legend.get_texts()[1].set_fontweight("bold")  # Set font weight for "Male"
legend.get_texts()[1].set_fontsize(12)  # Set font size for "Male"
plt.show()
```

From The above grapsh it is clear that most of the buyers are females and purchasing power of the females is greater than males.

State





```python
#Total number of orders from top 20 state
sales_gen=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders" , ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="State" , y="Orders" , data=sales_gen , width=0.7 )

plt.title("Total Number of orders from  10 state")


plt.show()
```

```python
#Total amount from top 10 state
sales_gen=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount" , ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="State" , y="Amount" , data=sales_gen , width=0.7 )

plt.title("Total amount  from  10 state")


plt.show()
```

From above graphs it is clear that most of the orders are from which state and in which state people spend amount the maximum

Product Category

```python
sns.set(rc={"figure.figsize":(30,9)})
ax = sns.countplot(x="Product_Category", data=df, width=0.9)
plt.title("No. of Orders  on the basis of Product Category " , size=10)

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Total amount spend on the basis of Product category
sales_gen=df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount" , ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="Product_Category" , y="Amount" , data=sales_gen , width=0.7 )
plt.title("Purchasing power (on the basis of Product Category) ")




plt.show()
```

From the above grapsh it is clear most of the sold products are Food clothing Mobile phones

```python

sales_gen=df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by="Orders" , ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="Product_ID" , y="Orders" , data=sales_gen , width=0.7 )
plt.title("Most sold product  (on the basis of Product ID) ")




plt.show()
```
